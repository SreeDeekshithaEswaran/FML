```python
#IMPORTING THE DATA OF TV AND SALES as x AND y
x=data["TV"].values
y=data["Sales"].values
```

```python
#checking datatypes of x and y
type(x),type(y)
```

```python
#checking the shape of x and y
print(x.shape,y.shape)
```

```python
#reshaping x and y
x=x.reshape(-1,1)
y=y.reshape(-1,1)
print(x.shape,y.shape)
```

```python
#splitting the train and test data
from sklearn.model_selection import train_test_split
train_x,test_x,train_y, test_y=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
#prearing and training model
from sklearn import linear_model
z=linear_model.LinearRegression() 
z.fit(train_x,train_y)
```

```python
#calculating mse,r2 score,mae
y_pred=(z.predict(test_x))
print(f'MAE = {mean_absolute_error(test_y, y_pred)}')
print(f'MSE = {mean_squared_error(test_y, y_pred)}')
print(f'RMSE = {np.sqrt(mean_squared_error(test_y, y_pred))}')
print(f'R_2 = {r2_score(test_y, y_pred)}')
```
